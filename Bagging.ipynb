```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# To ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Reading the CSV file and putting it into the 'df' object.
df = pd.read_csv(r'/content/credit-card-default (1).csv')  # Replace 'your_file.csv' with the actual file path or URL

# Printing the head of the DataFrame
df.head()
```

```python
df.info()
```

```python
# Importing train_test_split from sklearn library
from sklearn.model_selection import train_test_split
X = df.drop('defaulted',axis=1)
y = df['defaulted']
# Assuming you have already loaded your data into X and y

# Splitting the data into train and test with a test size of 30% and random state as 101
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Importing necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assuming you have already split your data into X_train, X_test, y_train, y_test

# Creating a pipeline with StandardScaler and DecisionTreeClassifier
pipeline = Pipeline([
    ('standardscaler', StandardScaler()),
    ('decisiontreeclassifier', DecisionTreeClassifier(random_state=1))
])

# Fitting the pipeline on the training data
pipeline.fit(X_train, y_train)

# Model scores on test and training data
y_test_pred = pipeline.predict(X_test)
y_train_pred = pipeline.predict(X_train)

test_score = accuracy_score(y_test, y_test_pred)
train_score = accuracy_score(y_train, y_train_pred)

print(f"Model test Score: {test_score:.3f}, Model training Score: {train_score:.3f}")
```

```python
# Importing necessary libraries
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Assuming you have already created the pipeline and split your data into X_train, X_test, y_train, y_test

# Instantiate the bagging classifier with base_estimator as the pipeline and random_state=1
bgclassifier = BaggingClassifier(base_estimator=pipeline, random_state=1)

# Fitting the model on the training data
bgclassifier.fit(X_train, y_train)

# Model scores on test and training data
y_test_pred = bgclassifier.predict(X_test)
y_train_pred = bgclassifier.predict(X_train)

test_score = accuracy_score(y_test, y_test_pred)
train_score = accuracy_score(y_train, y_train_pred)

print(f"Model test Score: {test_score:.3f}, Model training Score: {train_score:.3f}")
```

```python
# Importing necessary libraries
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Assuming you have already split your data into X_train, X_test, y_train, y_test

# A dictionary with base_estimator__max_depth from 1 to 20 with a step size of 5
# max_samples with [0.05, 0.1, 0.2, 0.5]
# max_features from 1 to 20 with a step size of 5
param_grid = {
    'base_estimator__max_depth': [1, 5, 10, 15, 20],
    'max_samples': [0.05, 0.1, 0.2, 0.5],
    'max_features': [1, 5, 10, 15, 20]
}

# Instantiate GridSearchCV with BaggingClassifier, base estimator as DecisionTreeClassifier, and random_state=1
gs = GridSearchCV(
    estimator=BaggingClassifier(base_estimator=DecisionTreeClassifier(), random_state=1),
    param_grid=param_grid,
    scoring='accuracy'
)

# Fit grid search on training data
gs.fit(X_train, y_train)

# Print the best parameters selected by GridSearchCV
print("Best Parameters:", gs.best_params_)
```

```python
# Print mean accuracy score on the final tuned BaggingClassifier
print("Optimal hyperparameter combination:", gs.best_params_)
print("Mean cross-validated training accuracy score:", gs.best_score_)

# Check BaggingClassifier again on optimized parameters
# Instantiate the bagging classifier with optimized parameters by grid search and random_state=1
bgclassifier = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(max_depth=gs.best_params_['base_estimator__max_depth']),
    max_features=gs.best_params_['max_features'],
    max_samples=gs.best_params_['max_samples'],
    random_state=1
)

# Fitting the above model on the training data
bgclassifier.fit(X_train, y_train)

# Model scores on test and training data
y_test_pred = bgclassifier.predict(X_test)
y_train_pred = bgclassifier.predict(X_train)

test_score = accuracy_score(y_test, y_test_pred)
train_score = accuracy_score(y_train, y_train_pred)

print("Model test Score:", test_score)
print("Model training Score:", train_score)
```

```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Assuming you have already split your data into X_train, X_test, y_train, y_test

# Running the random forest with default parameters, but random_state as 1
rfc = RandomForestClassifier(random_state=1)

# Fitting the model on the training data
rfc.fit(X_train, y_train)

# Making predictions
predictions = rfc.predict(X_test)

# Importing classification report and confusion matrix from sklearn metrics
# Let's check the report of our default model
print("Classification Report:")
print(classification_report(y_test, predictions))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
```

```python
# Importing necessary libraries
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Specify the number of folds for k-fold CV, which is 5
n_folds = 5

# Parameters to build the model; here, 'max_depth' with range(2, 20, 5)
parameters = {'max_depth': range(2, 20, 5)}

# Instantiate the Random Forest model with random_state=1
rf = RandomForestClassifier(random_state=1)

# Instantiate GridSearchCV with Random Forest, parameters, cv, and scoring as accuracy, and return_train_score=True
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=parameters,
    cv=KFold(n_splits=n_folds, shuffle=True, random_state=1),
    scoring='accuracy',
    return_train_score=True
)

# Fit the model on training data
grid_search.fit(X_train, y_train)

# Scores of GridSearch CV
scores = pd.DataFrame(grid_search.cv_results_)

# Convert scores into a DataFrame and print the head
print("Head of the Scores DataFrame:")
print(scores.head())
```

```python
import matplotlib.pyplot as plt

# Extracting the necessary columns from the scores DataFrame
max_depth_values = scores['param_max_depth'].astype(int)
mean_test_scores = scores['mean_test_score']
mean_train_scores = scores['mean_train_score']

# Plotting the accuracies with max_depth
plt.figure(figsize=(10, 6))
plt.plot(max_depth_values, mean_train_scores, label='Mean Train Score', marker='o')
plt.plot(max_depth_values, mean_test_scores, label='Mean Test Score', marker='o')

plt.title('Mean Train and Test Scores with Different max_depth Values')
plt.xlabel('max_depth')
plt.ylabel('Mean Accuracy Score')
plt.legend()
plt.grid(True)
plt.show()
```
