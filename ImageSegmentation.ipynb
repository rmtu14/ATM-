```python
import cv2
import matplotlib.pyplot as plt

# Load your image (replace 'your_image.jpg' with the actual file path)
image_path = '/content/WhatsApp Image 2023-11-28 at 10.26.26 PM.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Resize the image to be compatible with 8x8
resized_image = cv2.resize(image, (8, 8))

# Flatten the resized image
flattened_image = resized_image.flatten()

# Reshape the flattened image to 8x8
reshaped_image = flattened_image.reshape(8, 8)

# Plot the reshaped image
plt.figure(figsize=(8, 8))
plt.imshow(reshaped_image, cmap='gray')
plt.axis('off')  # Turn off axis
plt.show()

# Print the shape of the flattened image
print("Flattened Image Shape:", flattened_image.shape)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load your image (replace 'your_image.jpg' with the actual file path)
image_path = '/content/WhatsApp Image 2023-11-28 at 10.26.26 PM.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Otsu's method
ret, thresh1 = cv2.threshold(image, 0, 255, cv2.THRESH_OTSU)

# Display the original and thresholded images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(thresh1, cmap='gray')
plt.title('Otsu Thresholded Image')
plt.axis('off')

plt.show()

# Print the optimal threshold value and the return value
print("Optimal Threshold Value (ret):", ret)
print("Thresholded Image Shape:", thresh1.shape)
```

```python
import cv2
import matplotlib.pyplot as plt
from skimage import filters

# Replace 'your_image.jpg' with the actual file path of your image
image_path = '/content/WhatsApp Image 2023-11-28 at 10.26.26 PM.jpeg'

# Read the image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Sobel edge operator
sobel_edges = filters.sobel(image)

# Plot the Sobel edges
plt.figure(figsize=(8, 8))
plt.imshow(sobel_edges, cmap='viridis')  # You can choose a different colormap if needed
plt.axis('off')  # Turn off axis
plt.show()
```
